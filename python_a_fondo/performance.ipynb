# Performance

```python
import matplotlib.pyplot as plt
import numpy as np
import timeit
```

```python
def calcular_tiempos(func, num_elements=500):
    results = []
    for x in range(1, num_elements):
        lista_elementos = [str(i) for i in range(x)]
        results.append(timeit.timeit(lambda: func(lista_elementos), number=1000))
    return results


def print_grafica(tiempos):
    plt.plot(tiempos, 'b')
    plt.xlabel('Inputs')
    plt.ylabel('Steps')
    plt.show()
```

# Constante. O(c)

```python
def recuperar_elemento_lista(mi_lista):
    result = mi_lista[0]
    return result

tiempos = calcular_tiempos(recuperar_elemento_lista)
print_grafica(tiempos)
```

# Linear. O(n)

```python
def recuperar_elementos_lista(mi_lista):
    for elemento in mi_lista:
        result = elemento
    return result

tiempos = calcular_tiempos(recuperar_elementos_lista)
print_grafica(tiempos)
```

# Cuadrático. O(n^2) 

```python
def recuperar_elementos_lista(mi_lista):
    for elemento in mi_lista:
        for elemento_2 in mi_lista:
            result = elemento_2
    return result

tiempos = calcular_tiempos(recuperar_elementos_lista, num_elements=90)
print_grafica(tiempos)
```

# Exponencial. O(2^n)

```python
def calcular_tiempos_enteros(func, num_elements=500):
    result = []
    for element in range(num_elements):
        result.append(timeit.timeit(lambda: func(element), number=1000))
    return result


def fibonacci(n):
    if n in {0, 1}:
        return 1
    else:
        return fibonacci(n - 1) + fibonacci(n - 2)

tiempos = calcular_tiempos_enteros(fibonacci, num_elements=23)
print_grafica(tiempos)
```

## Ver si un elemento existe en una lista

```python
def dime_si_existe_un_elemento(mi_lista):
    result = "NO existe"
    if "10" in mi_lista:
        result = "Existe"

    return result

print("El elemento ", dime_si_existe_un_elemento(["7", "8", "9", "10"]))
print("El elemento ", dime_si_existe_un_elemento(["10", "7", "8", "9"]))
print("El elemento ", dime_si_existe_un_elemento(["7", "8", "9", "20", "21"]))
```

```python
tiempos = calcular_tiempos(dime_si_existe_un_elemento, num_elements=500)
print_grafica(tiempos)
```

```python
def dime_si_existe_un_elemento_alternativa(mi_lista):
    result = "NO existe"
    try:
        mi_lista.index("10")
        result = "Existe"
    except ValueError:
        pass
    return result

print("El elemento ", dime_si_existe_un_elemento_alternativa(["7", "8", "9", "10"]))
print("El elemento ", dime_si_existe_un_elemento_alternativa(["10", "7", "8", "9"]))
print("El elemento ", dime_si_existe_un_elemento_alternativa(["7", "8", "9", "20", "21"]))
print("El elemento ", dime_si_existe_un_elemento_alternativa([]))
```

```python
tiempos = calcular_tiempos(dime_si_existe_un_elemento_alternativa, num_elements=500)
print_grafica(tiempos)
```

```python
def dime_si_existe_un_elemento_mal(mi_lista):
    result = "NO existe"
    for element in mi_lista:
        if element == "10":
            result = "Existe"

    return result

tiempos = calcular_tiempos(dime_si_existe_un_elemento_mal, num_elements=500)
print_grafica(tiempos)
```

```python
from copy import copy

def eliminar_elementos(mi_lista):
    lista_copia = copy(mi_lista)
    for i in range(len(mi_lista)):
        lista_copia.pop()

tiempos = calcular_tiempos(eliminar_elementos, num_elements=150)
print_grafica(tiempos)
```

```python
def eliminar_elementos(mi_lista):
    lista_copia = copy(mi_lista)
    for i in range(len(lista_copia)):
        lista_copia.remove(str(i))

tiempos = calcular_tiempos(eliminar_elementos, num_elements=150)
print_grafica(tiempos)
```

```python
def aniadir_elementos_al_final(mi_lista):
    num_elements = len(mi_lista)
    nueva_lista = []
    for i in range(num_elements):
        nueva_lista.append(str(i))

tiempos = calcular_tiempos(aniadir_elementos_al_final)
print_grafica(tiempos)
```

```python
def aniadir_elementos_al_principio(mi_lista):
    num_elements = len(mi_lista)
    nueva_lista = []
    for i in range(num_elements):
        nueva_lista.insert(0, str(i))

tiempos = calcular_tiempos(aniadir_elementos_al_principio)
print_grafica(tiempos)
```

## Diccionarios

```python
def calcular_tiempos_diccionario(func, num_elements=500):
    results = []
    for x in range(1, num_elements):
        lista_elementos = {str(i): i for i in range(x)}
        results.append(timeit.timeit(lambda: func(lista_elementos), number=1000))
    return results
```

```python
def dime_si_existe_un_elemento(mi_diccionario):
    result = "NO existe"
    if "10" in mi_diccionario:
        result = "Existe"

    return result

print("El elemento ", dime_si_existe_un_elemento({"7": "10", "8": "9", "10": "50000"}))
print("El elemento ", dime_si_existe_un_elemento({"8": "9"}))
```

```python
tiempos = calcular_tiempos_diccionario(dime_si_existe_un_elemento, num_elements=10000)
print_grafica(tiempos)
```
