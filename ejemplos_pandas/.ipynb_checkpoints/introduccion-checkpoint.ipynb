# Trabajando con Datos

Datos descargables desde: https://www.kaggle.com/camnugent/sandp500

````python
df = pd.read_csv("all_stocks_5yr.csv")
df
````

````python
df.describe()
````

````python
df.dtypes
````

````python
df['Name'].unique()
````

````python
subdf = df[df["Name"] == "GOOGL"]
subdf
````

````python
subdf.plot(x="date", y="close")
````

````python
subdf[subdf["date"] > "2018-01-20"].plot(x="date", y="close")
````

````python
df[["date", "close"]]
````

````python
df[["date", "close"]].groupby(["date"]).agg({"close": [np.size, np.mean]})
````

````python
df[df["date"] == "2013-02-08"]["close"].count()
````

````python
df[df["date"] == "2013-02-08"]["close"].sum() / df[df["date"] == "2013-02-08"]["close"].count()
````

````python
df[["date", "close"]].groupby(["date"]).agg({"close": np.mean}).plot()
````

### Trabajando con fechas

````python
df.dtypes
````

````python
df["date"]
````

````python
pd.Timestamp(df["date"][0])
````

````python
df_dates = df.copy()
````

````python
%%time
df_dates['date'] = [pd.Timestamp(date) for date in df['date']]
df_dates
````

````python
df_dates.dtypes
````

````python
%%time
df_dates['date'] =  pd.to_datetime(df['date'], format='%Y-%m-%d')
df_dates
````

````python
df_dates.dtypes
````

````python
df_dates['week_day'] = [pd.Timestamp(date).day_name() for date in df['date']]
df_dates
````

````python
df_dates['week_day'] = df_dates['date'].map(lambda date: pd.Timestamp(date).day_name())
df_dates
````

````python
(pd.Timestamp(df['date'][0]) - pd.Timedelta("1 day")).day_name()
````

````python
(pd.Timestamp(df['date'][0]) + pd.offsets.BDay()).day_name()
````
